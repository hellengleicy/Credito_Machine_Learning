```python
#dados a serem tratados na base de treino
for i in features_test:
    print(i + '    :    ' + str(sum(data_train[i].isnull())))
```

```python
data_train.loc[data_train.credit_limit.isnull(),'credit_limit'] = Traincl
data_train.loc[data_train.reason.isnull(),'reason'] = Trainre#.item
data_train.loc[data_train.sign.isnull(),'sign'] = Trainsi#.item
data_train.loc[data_train.gender.isnull(),'gender'] = Trainge#.item
data_train.loc[data_train.facebook_profile.isnull(),'facebook_profile'] = Trainfa#.item
data_train.loc[data_train.job_name.isnull(),'job_name'] = Trainjo#.item
data_train.loc[data_train.ok_since.isnull(),'ok_since'] = Trainok
data_train.loc[data_train.n_bankruptcies.isnull(),'n_bankruptcies'] = Trainba
data_train.loc[data_train.n_defaulted_loans.isnull(),'n_defaulted_loans'] = Trainde
data_train.loc[data_train.n_issues.isnull(),'n_issues'] = Trainis
```

```python
Trainsc1 = '999'#data_train.score_1.mode()
Trainsc2 = '999'#data_train.score_2.mode()
Trainsc3 = data_train.score_3.median()
Trainri = data_train.risk_rate.median()
Trainam = data_train.amount_borrowed.mean()
Trainbo = data_train.borrowed_in_months.median()
Trainin = data_train.income.mean()
Trainst = '999'#data_train.state.mode()
Trainzi = '999'#data_train.zip.mode()
Trainch = '999'#data_train.channel.mode()
Trainre = '999'#data_train.real_state.mode()
Trainacc = data_train.n_accounts.mean()
```

```python
data_train.loc[data_train.score_1.isnull(),'score_1'] = Trainsc1#.item
data_train.loc[data_train.score_2.isnull(),'score_2'] =Trainsc2#.item
data_train.loc[data_train.score_3.isnull(),'score_3'] =Trainsc3
data_train.loc[data_train.risk_rate.isnull(),'risk_rate'] =Trainri
data_train.loc[data_train.amount_borrowed.isnull(),'amount_borrowed'] = Trainam
data_train.loc[data_train.borrowed_in_months.isnull(),'borrowed_in_months'] = Trainbo 
data_train.loc[data_train.income.isnull(),'income'] = Trainin 
data_train.loc[data_train.state.isnull(),'state'] = Trainst#.item 
data_train.loc[data_train.zip.isnull(),'zip'] = Trainzi#.item
data_train.loc[data_train.channel.isnull(),'channel'] =Trainch#.item
data_train.loc[data_train.real_state.isnull(),'real_state'] =Trainre#.item
data_train.loc[data_train.n_accounts.isnull(),'n_accounts'] =Trainacc
```

```python
print(sum(data_train.default.isnull()),data_train.shape)
```

```python
data_train = data_train.drop(data_train.loc[data_train.default.isnull()].index)
#data_train.loc[data_train.default.isnull(),'default'] = False
```

```python
#data_train.loc[data_train.default.isnull()]#,data_train.index]
#data_train.index
data_train.loc[data_train.default == True,['default']] = int(True)
data_train.loc[data_train.default == False,['default']] = int(False)
```

```python
data_train.default
```

```python
for a, b,c in [[data_train.score_1,'score_1',data_test.score_1],[data_train.score_2,'score_2',data_test.score_2],[data_train.sign,'sign',data_test.sign],[data_train.gender,'gender',data_test.gender], [data_train.facebook_profile,'facebook_profile',data_test.facebook_profile],[data_train.state,'state',data_test.state],[data_train.channel,'channel',data_test.channel],[data_train.real_state,'real_state',data_test.real_state]]:
    my_list = list(set(a))
    for i in my_list:
        #print(i)
        data_train.loc[a == i ,[b]] = int(my_list.index(i))
        data_test.loc[c == i ,[b]] = int(my_list.index(i))
    data_train = data_train.astype({b: int})
    data_test = data_test.astype({b: int})
```

```python
data_test.channel
```

```python
from sklearn.model_selection import train_test_split
#Neste ponto posso escolher minhas variÃ¡veis, tirando aquelas que julgar nÃ£o importantes
X = data_train[['score_1',
 'score_2',
 'score_3',
 'score_4',
 'score_5',
 'score_6',
 'risk_rate',
 'amount_borrowed',
 'borrowed_in_months',
 'credit_limit',
 #'reason',
 'income',
 'sign',
 'gender',
 'facebook_profile',
 'state',
 #'zip',
 'channel',
 #'job_name',
'real_state',
 'ok_since',
 'n_bankruptcies',
 'n_defaulted_loans',
 'n_accounts',
 'n_issues']]    
y = data_train.default
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
print(data_train.shape)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#from catboost import CatBoostRegressor
#categorical_features_indices = np.where(X.dtypes != np.float)[0]
```

```python
categorical_features_indices = [ 0,  1, 11, 12, 13, 14, 15, 16]
```

```python
train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=list(categorical_features_indices))
test_data = lgb.Dataset(X_test, label=y_test,reference=train_data)#,categorical_feature=list(categorical_features_indices))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

params = {
    'application': 'binary',
    'objective': 'binary',
    'metric': 'auc',
 #  'is_unbalance': 'true',
    'boosting': 'gbdt',
    'num_leaves': 50,
    'feature_fraction': 0.4,
    'bagging_fraction': 0.6,
    'bagging_freq': 20,
    'learning_rate': 0.01,
    'verbose': 0
}
#lf = lgb.LGBMRegressor(**params, random_state=42)
clf  = lgb.train(params,
                train_data,
                num_boost_round=100,
                valid_sets=test_data,
                early_stopping_rounds=5)


predictions = clf.predict(X_test, num_iteration=clf.best_iteration)
# eval
#print('The rmse of prediction is:', mean_squared_error(y_test, y_pred) ** 0.5)

#predictions = clf.predict(test_data,num_interation=clf.best_iteration)
auc = roc_auc_score(y_test, predictions)
print("auc_score: %.4f" % auc)

```

```python
data_test['prob'] = clf.predict(data_test.drop(columns =['ids','reason', 'zip', 'job_name']))
```

```python
df = data_test[['ids','prob']]
df.to_csv("submission014.csv",index=False)
```
